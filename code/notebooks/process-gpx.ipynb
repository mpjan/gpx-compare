```python
import pandas as pd
import gpxpy
import os
from pathlib import Path
from tqdm import tqdm
```

```python
GPX_FILE_PATH = '../../data/gpx/'
```

```python
def process_gpx_files(gpx_file_path=GPX_FILE_PATH):
    # Get list of GPX files
    gpx_dir = Path(gpx_file_path)
    gpx_files = list(gpx_dir.glob('*.gpx'))
    gpx_files.sort()

    # Create empty lists to store data
    file_names = []
    track_names = []

    # Process each GPX file
    for gpx_file in tqdm(gpx_files):
        with open(gpx_file, 'r', encoding='utf-8') as f:
            gpx = gpxpy.parse(f)
            
            file_names.append(gpx_file.name)
            track_names.append(gpx.name if gpx.name else None)
            if gpx.name is None:
                print(f"No track name found for {gpx_file.name}")

    # Create DataFrame
    df = pd.DataFrame({
        'file_name': file_names,
        'track_name': track_names
    })

    return df
```

```python
df = process_gpx_files()
```
